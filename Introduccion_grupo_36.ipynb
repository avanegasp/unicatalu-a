3 tipos de archivos de salida:
- Excel
- CSV
- Parquet
- Pickle (opcional)

```python
import pandas as pd
import time

# Suponiendo que 'airbnb' es tu dataframe de pandas

path_csv = '/content/drive/MyDrive/U de cataluña/grupo_36/raw/prueba.csv'
path_excel = '/content/drive/MyDrive/U de cataluña/grupo_36/raw/prueba.xlsx'
path_parquet = '/content/drive/MyDrive/U de cataluña/grupo_36/raw/prueba.parquet'

# Exportar a CSV
%time airbnb.to_csv(path_csv, index=False)

# Exportar a Excel
%time airbnb.to_excel(path_excel, index=False)

# Exportar a Parquet
%time airbnb.to_parquet(path_parquet)
```

```python
# Leer CSV
%time airbnb = pd.read_csv(path_csv,)

# Leer Excel
%time airbnb = pd.read_excel(path_excel,)

# Lee Parquet
%time airbnb = pd.read_parquet(path_parquet)
```
